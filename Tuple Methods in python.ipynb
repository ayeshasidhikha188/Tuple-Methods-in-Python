# Tuples
* Tuple is a collection of data separated by commas. 
* In some ways, a tuple is similar to a  list in terms of indexing, nested objects, and repetition but the main difference between both is tuple is immutable, unlike the list which is mutable.
* Tuples are used to store multiple items in a single variable.
* A tuple is a collection which is ordered and unchangeable.
* Tuples are written with round brackets.


```python
a=(1,2.2,3+3j,'ayesha',print,range)
print(a,type(a),id(a),sep='\n')
```

### Creating Python Tuples
* There are two ways by which a tuple can b created. They are as follows:
   * Using round brackets
   * Tuple Constructor
#### Create Tuple using Round Brackets ()
* To create a tuple we will use () operators

```python
a=(1,2.2,3+3j,'four',print,range)
print(a,type(a),id(a),sep='\n')
```

```python
c,d=('hello'),({1,100,99})
print(c,d) # () does not make substrings
```

#### using tuple constructor
* To create a tuple with a Tuple constructor, we will pass the elements as its parameters.

```python
b=tuple(range(5))
print(b)
```

```python
b=tuple('hello')
print(b) # same as list 
```

```python
c,d=tuple('hello'),tuple({1,100,99})
print(c,d) # tuple class makes substrings
```

## Accessing  elements from tuple
* Tuples are immutable, and usually, they contain a sequence of heterogeneous elements that are accessed via unpacking or indexing (or even by attribute in the case of named tuples). 
* **Note:In unpacking of tuple number of variables on the left-hand side should be equal to a number of values in given tuple a.** 
* To access values or elements in  tuple, we can use indexing or slicing. 
* Indexing starts from 0, which means the first element has an index of 0, the second element has an index of 1, and so on.
* We can  use negative indexing, which starts from the last element (-1), the second last element (-2),andsoon.
### Indexing in  tuple
* Indexing is a technique for accessing the elements of a tuple. There are various ways by which we can access an element of a tuple.
## Positive Index
* In the case of Positive Indexing, the first element of the list has the index number 0, and the last element of the list has the index number N-1, where N is the total number of elements in the tuple (size of the tuple).


```python
x= (50,70,30,20,90,10,50)
print(x,type(x),id(x),sep='\n')
print(x[0])
```

## Negative Index
* Index -1 represents the last element and -N represents the first element of the tuple, where N is the length of the tuple


```python
x= (50,70,30,20,90,10,50)
print(x,type(x),id(x),sep='\n')
print(x[-1])
```

## Slicing
* Syntax tuple[ Initial : End : IndexJump ]
* If t is a tuple, then the above expression returns the portion of the tuple from index Initial to index End, at a step size IndexJump
* It can be  both with positive indexes as well as negative indexes.


```python
x= (50,70,30,20,90,10,50)
print(x,type(x),id(x),sep='\n')
print(x[1:5])
```

## Length of  Tuple
* len() is used to get the length of the tuple.

```python
x= (50,70,30,20,90,10,50)
len(x)
```

## Concatenation of Tuples
* Concatenation of tuple is the process of joining two or more Tuples. 
* Concatenation is done by the use of + operator. 
* Concatenation of tuples is done always from the end of the original tuple. 
* Other arithmetic operations do not apply on Tuples. 
* Only the same datatypes can be combined with concatenation, an error arises if a list and a tuple are combined. 


```python
t1,t2=(0,1,2,3),('ayesha','sidhikha')
t3=t1+t2
print(t3)
```

### Methods in Tuple

* Returns the frequency of occurrence of a specified value

## 1. Count()
* The count() method of Tuple returns the number of times the given element appears in the tuple.
* Syntax:tuple.count(element)
* here the element is the element that is to be counted
##### Parameters: 
* element: is the item whose count is to be returned.
##### Returns: 
* Returns the count of how many times element occurs in the tuple.

```python
t=(1,2,2,4,2)
t.count(2)
```

## 2.  Index() 
* The Index() method returns the first occurrence of the given element from the tuple.
* Syntax:tuple.index(element, start, end)
##### Parameters:
* element: The element to be searched.
* start (Optional): The starting index from where the searching is started
* end (Optional): The ending index till where the searching is done
* **Note: This method raises a ValueError if the element is not found in the tuple.**


```python
t=(1,3,2,5,3,7)
t.index(3)
```
